```python
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#Top 15 ages in the 50m race (20+ races min) 
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count > 19').head(15)
```

```python
#Bottom 15 ages in the 50m race (20+ races min) 
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=True).query('count > 10').head(15)
```

```python
#Seasons for the data -> Slower in summer than winter?
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(
    lambda x: 'Winter' if x > 11 else
              ('Fall'   if x > 8  else
               ('Summer' if x > 5  else
                ('Spring' if x > 2 else 'Winter'))))
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```

```python
#Mean Athlete Average Speed by Race Season
import matplotlib.pyplot as plt
df_plot = (
    df3.groupby('race_season')['athlete_average_speed']
       .agg(['mean','count'])
       .sort_values('mean', ascending=False)
)
# Plot the mean speeds
plt.figure()
df_plot['mean'].plot(kind='bar')
plt.style.use('tableau-colorblind10')
plt.title('Mean Athlete Average Speed by Race Season')
plt.xlabel('Race Season')
plt.ylabel('Mean Athlete Average Speed')
plt.show()
```
